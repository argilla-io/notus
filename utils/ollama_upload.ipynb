# Ollama upload

Note that you first need to have [ollama installed and configured for importing models](https://github.com/jmorganca/ollama/blob/main/docs/import.md).

```python
user = "TheBloke"
model = "notus-7B-v1"
quantizations = [
    "Q2_K", 
    "Q3_K_L", "Q3_K_M", "Q3_K_S",
    "Q4_0", "Q4_K_M", "Q4_K_S",
    "Q5_0", "Q5_K_M", "Q5_K_S",
    "Q6_K",
    "Q8_0"
]
overwrite = False
```

```python
from pathlib import Path

model_path = Path("models").mkdir(parents=True, exist_ok=True)
```

```python
for quantization in quantizations:
    url = f"https://huggingface.co/{user}/{model}-GGUF/resolve/main/{model.replace('B', 'b')}.{quantization}.gguf"
    local_path = f"models/{model}.{quantization}"
    if not Path(local_path).exists() or overwrite:
        print(f"Downloading {local_path} from Hugging Face")
        !wget -P {local_path} {url}
    else:
        print(f"Skipping {local_path} as it already exists")
```

```python
for quantization in quantizations:
    local_path = f"models/{model}.{quantization}"
    model_name: str = f"{model}.{quantization}.gguf"
    model_file_path = f"{local_path}/Modelfile"
    with open(model_file_path, "w+") as f:
        lines = [
            f"FROM ./{model_name}",
            "",
            "PARAMETER mirostat 0",
            "PARAMETER mirostat_eta 0.1",
            "PARAMETER mirostat_tau 5.0",
            "PARAMETER num_ctx 2048",
            "PARAMETER num_gpu 1",
            "PARAMETER repeat_last_n 64",
            "PARAMETER repeat_penalty 1.1",
            "PARAMETER temperature 0.8",
            "PARAMETER seed 0",
            "PARAMETER tfs_z 1",
            "PARAMETER num_predict 128",
            "PARAMETER top_k 40",
            "PARAMETER top_p 0.9",
            "",
            "TEMPLATE \"<|system|>\\n{{ .System }}</s>\\n<|user|>\\n{{ .Prompt }}</s>\\n<|assistant|>\\n\"",
            "",
            "SYSTEM \"\"\"You are a helpful assistant. Below is an instruction that describes a task. Write a response that appropriately completes the request.\"\"\""
        ]
        f.write("\n".join(lines))
    notus_path = f"notus:{quantization.replace('Q', 'q')}"
    !ollama create {notus_path} -f {model_file_path}
    !ollama cp {notus_path} davidberenstein1957/{notus_path}
    !ollama push davidberenstein1957/{notus_path}
    
        
```
